STEP 0: SETUP

```python
import pandas as pd
import matplotlib.pyplot as plt
```

STEP 1:DATA GENERATION

```python
import pandas as pd
import numpy as np
np.random.seed(42)
n = 50
data = {
    'Gender' : np.random.choice(['Male','Female'],n),
    'City' : np.random.choice(['Delhi','Mumbai','Pune','Bangalore'],n),
    'Age' : np.random.randint(18,25,n),
    'Study_Hours': np.random.randint(1,8,n),
    'Marks' : np.random.randint(40,95,n),
    'Satisfaction': np.random.choice(['Low','Medium','High'],n)
}
df = pd.DataFrame(data)
df.head()
```

STEP 2:UNDERSTAND GENERATERD DATA

```python
df.info()
df.describe()
df.isnull().sum()
```

STEP 3: SCALES OF MEASURMENT

Nominal Scale(Gender,City)

```python
df['Gender'].value_counts()
df['City'].value_counts()
```

Ordinal Scale(Satisfaction Level)

```python
ordinal_map={'Low': 1,'Medium': 2,'High': 3}
df['Satisfaction_Code']=df['Satisfaction'].map(ordinal_map)
df[['Satisfaction','Satisfaction_Code']].head()
```

Interval Scale(Marks)

```python
df['Marks'].mean()
df['Marks'].median()
```

Ratio Scale(Age,Study Hours)

```python
df['Age'].mean()
df['Study_Hours'].min()
df['Study_Hours'].max()
```

STEP 4: DATA ORGANIZATION

```python
df_sorted = df.sort_values(by='Marks')
df_sorted.head()
```

STEP 5: FREQUENCY DISTRIBUTION TABLE

```python
bins = [40,50,60,70,80,90]
df['Marks_Class'] = pd.cut(df['Marks'],bins=bins)
freq_table = df['Marks_Class'].value_counts().sort_index()
freq_table
```

STEP :6 BAR CHART(NOMINAL DATA)

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

STEP 7: PIE CHART(PROPORTION)

```python
df['City'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('City Distribution')
plt.ylabel('')
plt.show()
```

STEP 8: HISTOGRAM(CONTINUOUS DATA)

```python
df['Marks'].plot(kind='hist')
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.show()
```

STEP 9: LINE GRAPH(RELATIONSHIP)

```python
plt.plot(df['Study_Hours'],df['Marks'], 'o')
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.title('Study Hours vs Marks')
plt.show()
```

STEP 10:BOX PLOT(OUTLIERS)

```python
df.boxplot(column='Marks')
plt.title('Box Plot Of Marks')
plt.show()
```

STEP 11:FREQUENCY POLYGON

```python
midpoints = [45,55,65,75,85]
plt.plot(midpoints,freq_table)
plt.xlabel('Class Midpoints')
plt.ylabel('Frequency')
plt.title('Frequency Polygon')
plt.show()
```

STEP 12: OGIVE(CUMULATIVE FREQUENCY)

```python
cum_freq = freq_table.cumsum()
plt.plot(midpoints,cum_freq)
plt.xlabel('Class Midpoints')
plt.ylabel('Cumulative Frequency')
plt.title('Ogive Curve')
plt.show()
```

STEP 13: MEASURES OF CENTRAL TENDANCY

```python
mean_marks = df['Marks'].mean()
median_marks = df['Marks'].median()
mode_marks = df['Marks'].mode()
mean_marks,median_marks,mode_marks
```

STEP 14: WEIGHTED MEAN

```python
marks = [80,70,90]
credits = [3,4,2]
weighted_mean = sum(m*c for m,c in zip(marks,credits)) / sum(credits)
weighted_mean
```

STEP 15: COMBINED MEAN
