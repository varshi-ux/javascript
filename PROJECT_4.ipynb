STEP 1: DataSet Creation(Hands-On)

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
n = 50
data = {
    'Study_Hours':np.random.randint(1,8,n),
    'Marks':np.random.randint(40,95,n)
}
df = pd.DataFrame(data)
df.head()
```

PART 1:MEASURES OF DISPERSION

1Ô∏è‚É£ Range

```
data_range = df['Marks'].max() - df['Marks'].min()
data_range
```

2Ô∏è‚É£Varience And Standard Deviation

```
varience = df['Marks'].var()
std_dev = df['Marks'].std()
varience, std_dev
```

3Ô∏è‚É£ Interquartile Range (QR)

```
Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

4Ô∏è‚É£ Coefficient Of Variation(CV)

```
cv = (std_dev/df['Marks'].mean())*100
cv
```

PART 2: Skewness & Kurtosis

5Ô∏è‚É£Skewness(Shape Of Data)

```
skewness = df['Marks'].skew()
skewness
```

6Ô∏è‚É£Kurtosis(Peakedness)

```
kurtosis = df['Marks'].kurtosis()
kurtosis
```

Visualization(Very Important)

```
plt.figure(figsize=(8,5))
sns.histplot(df['Marks'],kde=True)
plt.title('Marks Distribution')
plt.show()
```

PART 3: Correlation  Analysis

7Ô∏è‚É£Pearson Correlation

```
pearson_corr = df['Study_Hours'].corr(df['Marks'],method='pearson')
pearson_corr
```

8Ô∏è‚É£Spearman Rank Correlation

```
spearman_corr = df['Study_Hours'].corr(df['Marks'],method='spearman')
spearman_corr
```

üìäScatter Plot

```
plt.figure(figsize=(6,4))
plt.scatter(df['Study_Hours'],df['Marks'])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title('Study Hours vs Marks')
plt.show()
```

PART 4: Regression Analysis(Linear)

9Ô∏è‚É£Simple Linear Regression

```
from sklearn.linear_model import LinearRegression
x = df[['Study_Hours']]
y = df['Marks']
model = LinearRegression()
model.fit(x,y)
slope = model.coef_[0]
intercept = model.intercept_
slope, intercept
```

üîÆPrediction Example

```
predicted_marks = model.predict([[6]])
predicted_marks
```

üìàRegression Line Plot

```
plt.scatter(x, y)
plt.plot(x, model.predict(x))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression: Study Hours vs Marks")
plt.show()
```
